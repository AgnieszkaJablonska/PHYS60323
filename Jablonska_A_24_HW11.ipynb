```python

# your solution here
import numpy as np
#1. define mass, position & velocity
class Particle(object):
    def __init__(self, x, y, z, m, vx, vy, vz):
        self.x = x
        self.y = y
        self.z = z
        self.m = m
        self.vx = vx
        self.vy = vy
        self.vz = vz
#2. Kinetic Energy
    def Kinetic_energy(self):
        #print( "The kinetic energy of the particle is ", 1/2 * self.m * (self.vx**2 + self.vy**2 + self.vz**2), "[J]", )
        return 1/2 * self.m * (self.vx**2 + self.vy**2 + self.vz**2)
#3. Distance
    def Distance(self, other):
        return np.sqrt((self.x - other.x)**2 + (self.y - other.y)**2 + (self.z - other.z)**2)
        #print( "The distance between two particles is", np.sqrt((self.x - other.x)**2 + (self.y - other.y)**2 + (self.z - other.z)**2), "[m]",)
        
#4. new position x + v * t
    def New_Position(self, dt):
        self.x = self.x + self.vx * dt
        self.y = self.y + self.vy * dt
        self.z = self.z + self.vz * dt
        #print( "The updated position of the particle based on the current position and velocity is represented by the coordinates (x, y, z) = (", (self.x + self.vx * dt), ",", (self.y + self.vy * dt), ",", (self.z + self.vz * dt), ")") 

#5. Adding charge

class Charged_Particle(Particle):
    def __init__(self, x, y, z, m, vx, vy, vz, q):
        self.x = x
        self.y = y
        self.z = z
        self.m = m
        self.vx = vx
        self.vy = vy
        self.vz = vz
        self.q = q

#6. Examples
#If there are interactions distances
    def Find_Seperation(self,other):
            if p1.Distance(p2) <= 0.25:
                if (self.q < 0 and other.q > 0) or (self.q > 0 and other.q < 0):
                    print("The particles interact, combine & merge since the charges are opposite")   #Checkig if the charges are opposite (positive - negative)
                    self.q = 0
                    other.q = 0
                    self.vx = 0
                    self.vy = 0
                    self.vz = 0
                    other.vx = 0
                    other.vy = 0
                    other.vz = 0 # When the charges are opposite
                    return True
                else:  #When charges are the same (positive-positive, negative-negative)
                    self.vx = self.vx*(-1 * (self.q+other.q) * (self.m/other.m))
                    self.vy = self.vy*(-1 * (self.q+other.q) * (self.m/other.m))
                    self.vz = self.vz*(-1 * (self.q+other.q) * (self.m/other.m))
                    other.vx = other.vx*(-1 * (self.q+other.q) * (other.m/self.m))
                    other.vy = other.vy*(-1 * (self.q+other.q) * (other.m/self.m))
                    other.vz = other.vz*(-1 * (self.q+other.q) * (other.m/self.m))
                    return print("The particles interact & repel since the charges are the same ")
            else:
                return
```
